# EXERCISE 1

1. Dùng Gradient Descent với learning rate là 0.2 với function $f(x)=x^2$. <br>
Gradient của $f(x)$ w.r.t $x$:
$$
\frac{d f(x)}{d x}=2 x
$$
- update $x$ dùng gradient descent: 
$$
x_{\text {mới }}=x_{\text {cũ }}-\text { learning-rate } \times \frac{d f(x)}{d x}
$$
<br>

- Khởi tạo x = 6, learning rate = 0.2, iter 10 lần:

| x    | f(x)'|
| -------- | ------- |
| 3.6  | 0.2×2×6=2.4   |
| 2.16 | 0.2×2×3.6=1.44     |
| ...    | ...    |
| 0.03627    | ...   |

- Khởi tạo x = 6, learning rate = 2, iter 10 lần:

| x    | f(x)'|
| -------- | ------- |
| −18  | 2×2×6=24   |
| 54 | 2×2×(−18)=-72     |
| ...    | ...    |
| 354294    | ...   |

### SMALL LEARNING RATE VS BIG LEARNING RATE: CONVERGE VS DIVERGE

# EXERCISE 2

```python
import pandas as pd
from LinearReg import LinearRegression
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://raw.githubusercontent.com/nttuan8/DL_Tutorial/master/L1/data_linear.csv"
df = pd.read_csv(url)
```

```python
df[:5]
```

```python
df.shape
```

## SMALL LEARNING RATE

```python
X = df['Diện tích'].values.reshape(-1,1)
y = df['Giá'].values.reshape(-1,1)
```

```python
lin_reg_with_small_lr = LinearRegression(lr=0.00001, n_iter=10000)
```

```python
X_train, X_test, y_train, y_test = lin_reg_with_small_lr.train_test_split(X, y)
```

```python
X_train.shape
```

```python
lin_reg_with_small_lr.fit(X_train, y_train)
```

```python
y_pred = lin_reg_with_small_lr.predict(X_test)
```

```python
y_pred
```

```python
lin_reg_with_small_lr.visualize(X_train, y_train)
```

```python
plt.plot(lin_reg_with_small_lr.L_prevs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Iterations')
plt.show()
```

```python
print(
    f"""MSE: {lin_reg_with_small_lr.MSE(X_train, y_train)}
RMSE: {lin_reg_with_small_lr.RMSE(X_train, y_train)}
MAE: {lin_reg_with_small_lr.MAE(X_train, y_train)}
MAPE: {lin_reg_with_small_lr.MAPE(X_train, y_train)}
"""
)
```

#### BAD DATA + TOO SIMPLE MODEL = BAD PREDICTIONS
#### BUT MODEL CONVERGED

## BIG LERNING RATE

```python
lin_reg_with_big_lr = LinearRegression(lr=0.01, n_iter=10000)
```

```python
lin_reg_with_big_lr.fit(X_train, y_train)
```

```python
lin_reg_with_big_lr.w
```

```python
lin_reg_with_big_lr.predict(X_train)
```

#### MODEL DO NOT CONVERGE <=> DIVERGED
